## *Opdracht 6*: Meten en fitten, Leerdoel

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit 

arduino = np.array([0,357,393,431,447,470,493,525,543,560,567])
massa = np.array([0,25,50,75,100,125,150,175,200,225,250])/1000

def fit(x,a,b):
    return a*x**b
popt,pcov = curve_fit(fit,arduino,massa)


x_fit = np.linspace(min(arduino[1:]), max(arduino), 200)
y_fit = fit(x_fit, *popt)

plt.figure()
plt.errorbar(arduino,massa,5/1000,2,'k.',label='metingen')
plt.xscale('log')
plt.plot(x_fit,y_fit,label='fit lijn')
plt.xlabel('arduino meting')
plt.ylabel('massa [kg]')
plt.title('kalibreren van de arduino')
plt.legend()
plt.show()

print(popt)
```
